## Handling the missing values

```python
## using fillna function for the material column as it has more nan values
data['material_ref'].fillna('NA',inplace=True)
# material ref column after dealing with the nan values
data.isnull().sum()
```

```python
# we can drop the remaing nan value columns as it is less than 1% of the dataset 
data = data.dropna()
#dataset after dealing with all the null values
data.isnull().sum()
```

```python
## Dealing the dataset with correct format to develop a machine learning model
# changing the date's with correct date-time format
data['item_date'] = pd.to_datetime(data['item_date'])
data['delivery date'] =pd.to_datetime(data['delivery date'])

# changinng all the columns with float values into integer value
data['country'] = pd.to_numeric(data['country'].astype('Int64'))
data['application'] =pd.to_numeric(data['application'].astype('Int64'))

# changing reming columns into numeric values 
data['quantity tons'] = pd.to_numeric(data['quantity tons'], errors='coerce')
data['thickness'] = pd.to_numeric(data['thickness'],errors='coerce')
data['width'] =pd.to_numeric(data['width'],errors='coerce')
data['material_ref'] = data['material_ref'].str.lstrip('0')
data['product_ref'] = pd.to_numeric(data['product_ref'],errors='coerce')
data['selling_price'] =pd.to_numeric(data['selling_price'],errors='coerce')

# dataset after formating
data
```

## Model Building 

```python
# Encoding the categorical variables
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
# converting 
model = LabelEncoder()
data['status']=model.fit_transform(data['status'])

#OneHotEcoder
model1 =OneHotEncoder()
data_o = model1.fit_transform(data[['item type']])
dff =pd.get_dummies(data['item type'])
dff =dff.astype('int64')
```

```python
dff[['id','item_date','quantity tons','customer','country','status','application','thickness','width','material_ref','product_ref','delivery date','selling_price']]=data[['id','item_date','quantity tons','customer','country','status','application','thickness','width','material_ref','product_ref','delivery date','selling_price']]
```

```python
# dff=dff.drop(['item type'],axis=1)
dff.dtypes
```

## Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_curve, auc


# X =dff.drop(['item_date','delivery date'],axis=1,inplace=True)
X =dff.drop(['status'],axis=1)
y =dff['status']
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
model = LogisticRegression()
model.fit(x_train,y_train)
train_predict = model.predict(x_train)
test_predict = model.predict(x_test)
print('*********LOGISTIC REGRESSION*********')
print('*********Train*********')
print('Accuracy: ',accuracy_score(y_train,train_predict))
print('Precission: ',precision_score(y_train,train_predict,
                                           average='micro'))
print('Recall: ',recall_score(y_train,train_predict,
                                           average='micro'))
print('F1 Score: ',f1_score(y_train,train_predict,
                                           average='micro'))
print('********Test*********')
print('Accuracy: ',accuracy_score(y_test,test_predict))
print('Precission: ',precision_score(y_test,test_predict,
                                           average='micro'))
print('Recall: ',recall_score(y_test,test_predict,
                                           average='micro'))
print('F1 score: ',f1_score(y_test,test_predict,
                                           average='micro'))
```

## Extra Trees Classifier

```python
from sklearn.ensemble import ExtraTreesClassifier 

model = ExtraTreesClassifier()
model.fit(x_train,y_train)
train_predict = model.predict(x_train)
test_predict = model.predict(x_test)
print('*********Extra Trees Classifier*********')
print('*********Train*********')
print('Accuracy: ',accuracy_score(y_train,train_predict))
print('Precission: ',precision_score(y_train,train_predict,
                                           average='micro'))
print('Recall: ',recall_score(y_train,train_predict,
                                           average='micro'))
print('F1 Score: ',f1_score(y_train,train_predict,
                                           average='micro'))
print('********Test*********')
print('Accuracy: ',accuracy_score(y_test,test_predict))
print('Precission: ',precision_score(y_test,test_predict,
                                           average='micro'))
print('Recall: ',recall_score(y_test,test_predict,
                                           average='micro'))
print('F1 score: ',f1_score(y_test,test_predict,
                                           average='micro'))
```

## XGB Classifier

```python
import xgboost as xgb

model = xgb.XGBClassifier()
model.fit(x_train,y_train)
train_predict = model.predict(x_train)
test_predict = model.predict(x_test)
print('*********XGB Classifier*********')
print('*********Train*********')
print('Accuracy: ',accuracy_score(y_train,train_predict))
print('Precission: ',precision_score(y_train,train_predict,
                                           average='micro'))
print('Recall: ',recall_score(y_train,train_predict,
                                           average='micro'))
print('F1 Score: ',f1_score(y_train,train_predict,
                                           average='micro'))
print('********Test*********')
print('Accuracy: ',accuracy_score(y_test,test_predict))
print('Precission: ',precision_score(y_test,test_predict,
                                           average='micro'))
print('Recall: ',recall_score(y_test,test_predict,
                                           average='micro'))
print('F1 score: ',f1_score(y_test,test_predict,
                                           average='micro'))
```
